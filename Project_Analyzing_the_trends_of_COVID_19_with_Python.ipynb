# Problem Statement:
Given data about COVID-19 patients, write code to visualize the impact and
analyze the trend of rate of infection and recovery as well as make predictions
about the number of cases expected a week in future based on the current
trends.

(Use plotly (visualization library) to create interactive visualizations)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly
import plotly.express as ex
import plotly.graph_objects as go
```

```python
df=pd.read_csv('/content/covid_19_clean_complete.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df_new=df.groupby('Date')[['Confirmed','Deaths','Recovered','Active']].sum().reset_index()
```

```python
df_new=df_new.sort_values(by='Date')
```

```python
df_new.head()
```

Lets visualize the impact and analyze the trend of rate of infection and recovery

```python
df_new['Infection_Rate']=df_new['Confirmed'].pct_change().fillna(0)*100
df_new['Recovery_Rate']=df_new['Recovered'].pct_change().fillna(0)*100
```

```python
df_new.head()
```

```python
#Trend of Confirmed, Recovered, Active
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_new['Date'], y=df_new['Confirmed'], mode='lines+markers', name='Confirmed'))
fig.add_trace(go.Scatter(x=df_new['Date'], y=df_new['Recovered'], mode='lines+markers', name='Recovered'))
fig.add_trace(go.Scatter(x=df_new['Date'], y=df_new['Active'], mode='lines+markers', name='Active'))

fig.update_layout(title='COVID-19 Case Trends',
                  xaxis_title='Date',
                  yaxis_title='Number of Cases',
                  template='plotly_dark')
fig.show()
```

```python
fig.write_html(
    "covid_case_trends.html",
    include_plotlyjs="cdn",
    full_html=True
)
```

```python
#Rate of Infection and Recovery
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_new['Date'], y=df_new['Infection_Rate'], mode='lines+markers', name='Infection Rate (%)'))
fig.add_trace(go.Scatter(x=df_new['Date'], y=df_new['Recovery_Rate'], mode='lines+markers', name='Recovery Rate (%)'))

fig.update_layout(title='Infection and Recovery Rates Over Time',
                  xaxis_title='Date',
                  yaxis_title='Rate (%)',
                  template='plotly_dark')
fig.show()
```

```python
fig.write_html(
    "Infection_and_Recovery_Rates_Over_Time.html",
    include_plotlyjs="cdn",
    full_html=True
)
```

Lets make predictions about the number of cases expected a week in future based on the current trends.

(Use Facebook prophet library to make time series models.
And Visualize the prediction by combining these technologies)

```python
df_new.head()
```

```python
data=df_new.drop(columns=['Deaths','Recovered','Active','Infection_Rate','Recovery_Rate'])
```

```python
data.head()
```

```python
#Forecast with Facebook Prophet
from prophet import Prophet
```

```python
# Prophet requires 'ds' and 'y' column names
df_prophet = data.rename(columns={'Date': 'ds', 'Confirmed': 'y'})
```

```python
df_prophet.head(2)
```

```python
# Initialize and fit the model
model = Prophet()
model.fit(df_prophet)
```

```python
# Create future dates  7 days ahead
future = model.make_future_dataframe(periods=7)
```

```python
# Predict
forecast = model.predict(future)
```

```python
#Visualize Forecast with Plotly

fig = go.Figure()

fig.add_trace(go.Scatter(x=df_prophet['ds'], y=df_prophet['y'],mode='lines+markers', name='Actual'))
fig.add_trace(go.Scatter(x=forecast['ds'], y=forecast['yhat'],mode='lines', name='Forecast'))

fig.update_layout(title='Forecast of COVID-19 Confirmed Cases (Next 7 Days)',
                  xaxis_title='Date',
                  yaxis_title='Confirmed Cases',
                  template='plotly_dark')
fig.show()
```

```python
fig.write_html(
    "Forecast_COVID_Confirmed_Cases_(Next7_Days).html",
    include_plotlyjs="cdn",
    full_html=True
)
```
